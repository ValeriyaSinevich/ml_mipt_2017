### MSE

```python
result = minimize(lambda args: ((args[0] * x + args[1] - y)**2).sum(), [1, 0.])
```

```python
noise_2 = np.random.normal(0, 0.2**0.5, size=75)
```

```python
x_2 = np.random.uniform(0, 100, size=75)
```

```python
y_2 = -1 + noise_2
```

```python
y = np.concatenate((y, y_2))
```

```python
x = np.concatenate((x, x_2))
```

```python
pairs = zip(x, y)
```

```python
np.random.shuffle(pairs)
```

```python
x, y = zip(*pairs)
```

```python
x = np.array(x)
y = np.array(y)
```

```python
result_2 = minimize(lambda args: ((args[0] * np.array(x) + args[1] - np.array(y))**2).sum(), [1., 0.])
```

```python
print result_2.x
```

```python
colors = ListedColormap(['red', 'yellow'])

pyplot.figure(figsize(8, 8))
pyplot.plot(x, result_2.x[0] * x + result_2.x[1])
pyplot.scatter(x, y, s=2,c="y")
```

```python
result_3 = minimize(lambda args: (abs(args[0] * x + args[1] - y)).sum(), [1., 0.])
```

Я нашла ошибку! после zip не преобразовывала обратно к np.array, из-за этого args[0] * x сэмплировал x args[0] раз, вместо поэлементного умножения. Теперь все верно, и с помощью MAE модель верно оценила параметры.
