```python
def clean_data(data):
    for i in data.index.values:
#         print(data.loc[i])
        if pd.isnull(data.loc[i].values.any()):
            data = data.drop(i, axis=0)
    return data
```

```python
train = clean_data(train)
```

```python
def process_data(data):
    good_train = pd.DataFrame()
    created = False
    indices = []
    for i in data.index.values:
        t = data.loc[i]
        w = int(t['week'])
        y = int(t['year'])
        iid = int(t['item_id'])
        filtered = data.query('item_id == {}'.format(iid)).query('year == {}'.format(y))
        filtered= filtered[filtered.apply(lambda x: x['week'] - x['shift'] == w, axis=1)]
        if filtered.shape[0] > 0:
            if (filtered.shape[0] > 1):
                filtered = filtered[0:1]
            indices.append(i)
            new_row = filtered
#             new_row = new_row.reset_index([i]).drop('index', axis=1)
#             print(new_row)
            new_row = new_row.drop('shift', axis=1)
            new_row['week'] = w
            new_row['y'] = y
#             new_row = new_row.drop(new_row.columns[0], axis=1)
            if created:
                good_train = pd.concat((good_train, new_row))
            else:
                good_train = new_row
                created = True
    good_train = good_train.head().reset_index().drop('index', axis=1)
    return good_train, np.array(indices)
```

```python
def process_data(data):
    good_train = pd.DataFrame()
    created = False
    indices = []
    for i in range(data.shape[0]):
        t = data[i:i+1]
        w = int(t['week'])
        y = int(t['year'])
        iid = int(t['item_id'])
        filtered = data.query('item_id == {}'.format(iid)).query('year == {}'.format(y))
        filtered= filtered[filtered.apply(lambda x: x['week'] - x['shift'] == w, axis=1)]
        if filtered.shape[0] > 0:
            if (filtered.shape[0] > 1):
                filtered = filtered[0:1]
            indices.append(i)
            new_row = filtered
            new_row = new_row.drop('shift', axis=1)
            new_row['week'] = w
            new_row = new_row.drop(new_row.columns[0:1], axis=1)
            if created:
                good_train = pd.concat((good_train, new_row))
            else:
                good_train = new_row
                created = True
    return good_train, indices
```

```python
def combine_solutions(cool_reg, simple_reg, test):
    cool_test = process_data(test)
    y = simple_reg.predict(test)
    cool_y, indices = cool_reg.predict(cool_test)
    for i, ind in enumerate(indices):
        y[ind] = cool_y[i]
    return y
```
